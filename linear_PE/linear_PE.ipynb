```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def positional_encoding(seq_len, d_model):
    """
    Compute the positional encoding for a given sequence length and model dimensionality.
    
    Args:
    seq_len (int): Length of the sequence.
    d_model (int): Dimension of the model.
    
    Returns:
    np.ndarray: Positional encoding matrix of shape (seq_len, d_model).
    """
    PE = np.zeros((seq_len, d_model))
    
    for pos in range(seq_len):
        for i in range(d_model // 2):
            angle_rate = pos / (10000 ** (2 * i / d_model))
            PE[pos, 2 * i] = np.sin(angle_rate)
            PE[pos, 2 * i + 1] = np.cos(angle_rate)
    
    return PE

# Generate the positional encoding
seq_len = 27
d_model = 10
pos_encoding = positional_encoding(seq_len, d_model)

# # Separate even and odd positions
# even_positions = pos_encoding[:, ::2]
# odd_positions = pos_encoding[:, 1::2]

# plt.figure(figsize=(15, 7))
# for i in range(even_positions.shape[1]):
#     plt.plot(range(seq_len), even_positions[:, i], label=f'Dimension {2*i}')
# plt.title('Positional Encoding - Even Positions')
# plt.xlabel('Position')
# plt.ylabel('Encoding Value')
# plt.legend(loc='upper right')
# plt.show()

# # Plot odd positions
# plt.figure(figsize=(15, 7))
# for i in range(odd_positions.shape[1]):
#     plt.plot(range(seq_len), odd_positions[:, i], label=f'Dimension {2*i+1}')
# plt.title('Positional Encoding - Odd Positions')
# plt.xlabel('Position')
# plt.ylabel('Encoding Value')
# plt.legend(loc='upper right')
# plt.show()
```
